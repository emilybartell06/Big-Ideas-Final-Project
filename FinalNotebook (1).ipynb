**Sample Code:** Here is a sample of our code that we used to analyze the different datasets. This code deals with the firearm seizures dataset.

```python
import pandas as pd

import matplotlib.pyplot as plt

# set of 10 neighborhoods the group is choosing between
neighborhoods = ["Mount Washington", "East Hills", "Fairywood", "Central Business District", "East Liberty", "Brighton Heights", "Shadyside", "Hazelwood", "South Side Flats", "Squirrel Hill North", "Squirrel Hill South"]

# when I originally downloaded the csv it had a different name, but I changed the name to make it more readable when looking over my code.
df = pd.read_csv('firearms.csv')
```

```python
seizurecount = {neighborhood: 0 for neighborhood in neighborhoods}
for index, row in df.iterrows():
    neighborhood = row["neighborhood"]
    if neighborhood in neighborhoods:
        seizurecount[neighborhood] += row['total_count']

print("Number of gun seizures in each neighborhood:")
print(seizurecount)
```

```python
# Making bar graph
plt.bar(seizurecount.keys(), seizurecount.values())
plt.xlabel('Neighborhood')
plt.ylabel('Number of Firearm Seizures', rotation=90)
plt.title('Amount of Firearm Seizures per Neighborhood')
plt.xticks(rotation=70, size=8)
plt.show()
```
