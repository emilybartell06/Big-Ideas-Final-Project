```python
import pandas as pd
data = pd.read_csv('pps_schoolsapr2019publish.csv')

zipcodes = {"Mount Washington": [15211], "East Hills": [15221], "Fairywood": [15205], "Downtown": [15222, 15219], "East Liberty": [15206], "Brighton Heights": [15212], "Shadyside": [15232], "Hazelwood": [15207], "South Side Flats": [15203], "Squirrel Hill": [15217]}

print("__________________________________________")
print("Number of public schools:\n----")
for neighborhood, zips in zipcodes.items():
    school_count = data[data['zip'].isin(zips)].shape[0]
    print(f"{neighborhood}: {school_count} public schools")

print("__________________________________________")
print("Number of private schools:\n----")

data2 = pd.read_csv('25d92d95-f5f7-426b-8f97-97f8af4c49bd.csv')

for neighborhood, zips in zipcodes.items():
    school_count = data2[data2['location_zip_code'].isin(zips)].shape[0]
    print(f"{neighborhood}: {school_count} private schools")
```

A neighborhood's number of schools is an important submetric for deciding its livability because it shows how available education is in the area, which is essential to family and community growth. A higher number of public and private schools can give easier access and greater quality educational opportunities for children living in the area. One way to better understand how a neighborhood meets the educational needs of its residents, which is a part of its overall livability, is to look at the number of schools in the area.

The neighborhood with the highest number of schools is Brighton Heights, followed closely by East Hills. This indicates that according to this submetric, these neighborhoods would be the "best" to live in.
